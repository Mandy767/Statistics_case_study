## **Paired Sample-sign Test**

H0 : There is no significant difference between 2022 and 2023 runs scored

H1 : There is a significant difference between 2022 and 2023 runs scored

alpha=5%

```python
import pandas as pd
from scipy import stats


data = pd.read_csv('./data/runs_scored_2022_2023.csv')

data['Difference'] = data['2022 Runs Scored'] - data['2023 Runs Scored']

sign_array=[]

for i in data['Difference']:
    if(i<0):
        sign_array.append('-')
    elif(i>0):
        sign_array.append('+')
    else:
        sign_array.append('0')

print(sign_array)

positive_diff = sum(data['Difference'] > 0)
negative_diff = sum(data['Difference'] < 0)



test_val=min(positive_diff,negative_diff)


print("Number of positive differences:", positive_diff)
print("Number of negative differences:", negative_diff)
print("test Value",test_val)

critical_val=1 # alpha=5% and n=10

if(test_val>critical_val):
    print('Accept H0')
else:
    print('Reject H0,Accept H1')
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.bar(data.index, data['Difference'], color=['green' if val > 0 else 'red' if val < 0 else 'blue' for val in data['Difference']])
plt.xlabel('Index')
plt.ylabel('Difference in Runs Scored (2022 - 2023)')
plt.title('Differences in Runs Scored between 2022 and 2023')
plt.xticks(data.index)
plt.axhline(0, color='black', linewidth=0.8)
plt.show()
```
