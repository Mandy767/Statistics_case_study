## **GOODNESS OF FIT TEST**
The problem revolves around analyzing the distribution of wicket types in cricket matches involving the Royal Challengers Bangalore (RCB) team. The dataset contains the frequency of different wicket types (such as bowled, caught, lbw, etc.) in RCB matches

Null Hypothesis (H0): Data is good

Alternative Hypothesis (H1): Data is not good

```python
import numpy as np


RCB_data = [845, 272, 118, 80, 0, 36, 31, 2]

sum=0
for i in RCB_data:
    sum+=i

expected=[]

for i in range(8):
    expected.append(sum/8)

chi_sq_cal=0

for i in range(8):
    chi_sq_cal+=((RCB_data[i]-expected[i])**2/expected[i])

print(chi_sq_cal)


chi_sq_tab=12.017  #alpha=0.10	df=(8-1)=7

if(chi_sq_cal>chi_sq_tab):
    print("Reject H0")
else:
    print("Accept H0")
```

```python
import matplotlib.pyplot as plt


wicket_types = ['Caught', 'Bowled', 'Run Out', 'LBW', 'Retired Hurt', 'Stumped', 'Caught and Bowled', 'Hit Wicket']

plt.figure(figsize=(10, 6))
plt.plot(wicket_types, RCB_data, marker='o', linestyle='-', color='blue', label='Observed')
plt.plot(wicket_types, expected, marker='o', linestyle='-', color='red', label='Expected')


plt.xlabel('Wicket Types')
plt.ylabel('Frequency')
plt.title('Observed vs Expected Wicket Frequencies for Royal Challengers Bangalore')
plt.legend()


plt.xticks(rotation=45, ha='right')

plt.grid(True)


plt.tight_layout()
plt.show()

```
